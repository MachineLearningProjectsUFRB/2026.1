```python
import ollama
```

```python
models_dict = ollama.list()
```

```python
model_name = models_dict.models[11].model #gemma3:4b
print(model_name)
```

```python
# 2. Criar o prompt estruturado com clareza
prompt_estruturado = """
[PERSONA]
Atue como um Especialista em Marketing Digital e Redator Sênior focado em conversão.

[CONTEXTO]
Estamos lançando um aplicativo móvel de meditação e foco guiado chamado "MenteSerena". O público-alvo são jovens profissionais de 25 a 40 anos que sofrem com estresse no trabalho e falta de tempo.

[TAREFA]
Escreva uma sequência de 3 e-mails curtos de boas-vindas para novos usuários que acabaram de baixar o aplicativo. 

[REGRAS]
- O tom deve ser acolhedor, profissional e motivador.
- Cada e-mail deve focar em um benefício único (Ex: Redução de ansiedade, Foco no trabalho, Melhoria do sono).
- Inclua um Call to Action (CTA) claro no final de cada e-mail.

[FORMATO DE SAÍDA]
Formate a resposta usando Markdown. Use títulos (##) para separar cada e-mail, e negrito (**) para destacar as linhas de assunto.
"""

# 3. Enviar a requisição para o Ollama
print("Processando prompt estruturado...\n")
response = ollama.generate(model=model_name, prompt=prompt_estruturado)

# 4. Exibir o resultado formatado
print(response['response'])
```
