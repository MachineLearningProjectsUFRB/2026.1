```python
import os
import requests
import pandas as pd

# 1. Configurações Gerais
OUTPUT_DIR = "dados_censo_2017"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# URLs das Tabelas da API SIDRA
URL_TABELA_6955 = "https://apisidra.ibge.gov.br/values/t/6955/n6/2909802/v/9504,10081/p/all/c829/allxt/c227/4930/c218/allxt/c12517/111521"
URL_TABELA_6957 = "https://apisidra.ibge.gov.br/values/t/6957/n6/2909802/v/10084/p/all/c829/allxt/c226/4885/c218/allxt/c12517/111519"
URL_TABELA_6759 = "https://apisidra.ibge.gov.br/values/t/6759/n6/2909802/v/all/p/all/c829/allxt/c830/allxt/c218/allxt/c12517/111519,111521/c12567/allxt/c220/allxt"


def obter_dataframe_sidra(url, nome_tabela):
    print(f"Buscando dados da Tabela {nome_tabela}...")
    try:
        # Realiza a requisição com um tempo limite (timeout) de 120 segundos
        response = requests.get(url, timeout=120)
        if response.status_code != 200:
            print(f"Erro ao acessar a tabela {nome_tabela}: HTTP {response.status_code}")
            return pd.DataFrame()

        data = response.json()

        # O SIDRA retorna o cabeçalho descritivo na primeira linha [0] e os dados nas demais [1:]
        header = data[0]
        rows = data[1:]

        # Cria o DataFrame e renomeia as colunas usando a primeira linha retornada
        df = pd.DataFrame(rows)
        df.rename(columns=header, inplace=True)
        return df
    except Exception as e:
        print(f"Exceção ao buscar a tabela {nome_tabela}: {e}")
        return pd.DataFrame()


# 2. Requisição e Salvamento da Tabela 6955
df_6955 = obter_dataframe_sidra(URL_TABELA_6955, "6955")

if not df_6955.empty:
    caminho_6955 = os.path.join(OUTPUT_DIR, "tabela_6955.csv")
    df_6955.to_csv(caminho_6955, index=False, encoding="utf-8-sig")
    print(f"✓ Salvo com sucesso: {caminho_6955}")
    print("\nVisualização das primeiras linhas da Tabela 6955:")
    print(df_6955.head())
else:
    print("\nErro: Não foi possível obter ou salvar os dados da Tabela 6955.")

print("-" * 50)

# 3. Requisição e Salvamento da Tabela 6957
df_6957 = obter_dataframe_sidra(URL_TABELA_6957, "6957")

if not df_6957.empty:
    caminho_6957 = os.path.join(OUTPUT_DIR, "tabela_6957.csv")
    df_6957.to_csv(caminho_6957, index=False, encoding="utf-8-sig")
    print(f"✓ Salvo com sucesso: {caminho_6957}")
    print("\nVisualização das primeiras linhas da Tabela 6957:")
    print(df_6957.head())
else:
    print("\nErro: Não foi possível obter ou salvar os dados da Tabela 6957.")

print("-" * 50)

# 4. Requisição e Salvamento da Tabela 6759
df_6759 = obter_dataframe_sidra(URL_TABELA_6759, "6759")

if not df_6759.empty:
    caminho_6759 = os.path.join(OUTPUT_DIR, "tabela_6759.csv")
    df_6759.to_csv(caminho_6759, index=False, encoding="utf-8-sig")
    print(f"✓ Salvo com sucesso: {caminho_6759}")
    print("\nVisualização das primeiras linhas da Tabela 6759:")
    print(df_6759.head())
else:
    print("\nErro: Não foi possível obter ou salvar os dados da Tabela 6759.")
```
