```python
import os
import shutil

src_dirs = ["train", "valid"]
dst_dir = "data"

os.makedirs(dst_dir, exist_ok=True)

for src in src_dirs:
    if not os.path.isdir(src):
        print(f"Diretório não encontrado: {src}")
        continue

    for root, _, files in os.walk(src):
        rel_path = os.path.relpath(root, src)
        target_root = os.path.join(dst_dir, rel_path) if rel_path != "." else dst_dir
        os.makedirs(target_root, exist_ok=True)

        for filename in files:
            src_path = os.path.join(root, filename)
            dst_path = os.path.join(target_root, filename)

            if os.path.exists(dst_path):
                base, ext = os.path.splitext(filename)
                i = 1
                while True:
                    new_name = f"{base}_{i}{ext}"
                    new_dst_path = os.path.join(target_root, new_name)
                    if not os.path.exists(new_dst_path):
                        dst_path = new_dst_path
                        break
                    i += 1

            shutil.copy2(src_path, dst_path)
```

```python
image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff', '.webp'}
image_count = 0

for _, _, files in os.walk(dst_dir):
    for filename in files:
        if os.path.splitext(filename)[1].lower() in image_extensions:
            image_count += 1

print(f"Total de imagens em {dst_dir}: {image_count}")
```
