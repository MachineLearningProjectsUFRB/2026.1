```python
import ollama
```

```python
models_dict = ollama.list()

i = 0
for item in models_dict.models:
    print(f"{i} - {item.model}")
    i = i + 1
```

```python
model_name = models_dict.models[10].model #gemma3:4b
print(model_name)

response = ollama.chat(
    model=model_name,
    messages=[
        {
            'role': 'user',
            'content': 'Porquê o céu é azul? responda em duas sentenças.',
        },
    ],
)

print(response['message']['content'])
```
